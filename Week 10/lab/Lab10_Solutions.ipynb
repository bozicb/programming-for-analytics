# Week 10 – Seaborn Lab Solutions
Solutions for all tasks using `malazan_visualisation_big_dataset.csv`.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style='whitegrid')
df = pd.read_csv('malazan_visualisation_big_dataset.csv')
df.head()
```

## Task 1 – Scatterplot

```python
sns.scatterplot(data=df, x='Magic_Aptitude', y='Combat_Skill', hue='Army')
plt.show()
```

## Task 2 – KDE Plot

```python
sns.kdeplot(data=df, x='Morale', fill=True)
plt.show()
```
