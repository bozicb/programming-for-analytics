# Week 9 Lab – Solutions
Malazan NumPy + Preprocessing

```python
import pandas as pd
import numpy as np
df = pd.read_csv('malazan_characters_stats.csv')
df.head()
```

## Task 1 – Preprocessing

```python
# Strip and title-case character and army names
df['Character'] = df['Character'].astype(str).str.strip().str.title()
df['Army'] = df['Army'].astype(str).str.strip().str.title()

# Fill missing numeric values
df = df.fillna(df.mean(numeric_only=True))
df.head()
```

## Task 2 – NumPy Arrays

```python
strength = df['Strength'].to_numpy()
magic = df['Magic_Power'].to_numpy()
morale = df['Morale'].to_numpy()
battles = df['Battles_Fought'].to_numpy()
victories = df['Victories'].to_numpy()
strength, strength.shape
```

## Task 3 – Vectorised Computations

```python
win_rate = victories / battles * 100
power_index = (strength + magic + morale) / 3
df['Win_Rate'] = win_rate
df['Power_Index'] = power_index
df.head()
```

## Task 4 – Correlation Analysis

```python
corr_strength_victories = np.corrcoef(strength, victories)
corr_magic_morale = np.corrcoef(magic, morale)
corr_strength_victories, corr_magic_morale
```

## Task 5 – Ranking Champions

```python
ranking = df.sort_values(by='Power_Index', ascending=False)
ranking[['Character','Power_Index','Win_Rate']].head(3)
```
