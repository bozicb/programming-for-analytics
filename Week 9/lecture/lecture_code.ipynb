```python
# Starting HP (adjust as you like)
rake_hp = 300
icarium_hp = 300

# Random damage ranges
# Rake hits harder but less often
rake_damage_range = (20, 45)
# Icarium hits extremely hard but unpredictably
icarium_damage_range = (15, 60)

round_num = 1

while rake_hp > 0 and icarium_hp > 0:
    print(f"\n--- Round {round_num} ---")

    # Rake attacks Icarium
    rake_damage = np.random.randint(*rake_damage_range) # asterisk unpacks the tuple into two params
    icarium_hp -= rake_damage
    print(f"Rake deals {rake_damage} damage. Icarium HP: {max(icarium_hp, 0)}")

    if icarium_hp <= 0:
        break

    # Icarium attacks Rake
    icarium_damage = np.random.randint(*icarium_damage_range)
    rake_hp -= icarium_damage
    print(f"Icarium deals {icarium_damage} damage. Rake HP: {max(rake_hp, 0)}")

    round_num += 1

# Outcome
print("\n===== FINAL RESULT =====")
if rake_hp <= 0 and icarium_hp <= 0:
    print("Both warriors fall â€” a catastrophic draw.")
elif rake_hp <= 0:
    print("Icarium wins! Rake is defeated.")
else:
    print("Rake wins! Icarium is defeated.")
```
