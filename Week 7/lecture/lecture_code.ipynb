To make this more realistic we need to correlate height and weight when creating random values. Therefore, use `np.random.multivariate_normal` with covariance matrix:

`variance = std^2` 

`covariance = r * std_height * std_weight`

`covariance matrix = [[var_height, cov,], [cov, var_weight]]`
                      
`var_height = 7 * 7 = 49`

`var_weight = 5 * 5 = 25`

`cov = .6 * 7 * 5 = 21`

```python
hw = np.random.multivariate_normal([170, 80.7], [[49, 21],[21, 25]], 5000)
plt.scatter(hw[:,0], hw[:,1])
```

```python
import numpy as np
import pandas as pd

# Number of cadre mages
n = 80

# Income in gold falcons (per year)
mean_income_gold = 1.5      # average annual income in gold
std_income_gold = 0.4       # standard deviation
mean_income_silver = 52 * 5 + 52 * 6
std_income_silver = 171

# Exchange rate
silver_per_gold = 20 

# Generate a normal distribution of income (gold falcons)
gold_income = np.random.normal(mean_income_gold, std_income_gold, n)
silver_income = np.random.normal(mean_income_silver, std_income_silver, n)

# Prevent negative/zero income
gold_income = np.clip(gold_income, 0.2, None)
silver_income = np.clip(silver_income, 100, None)

# Convert to silver falcons
overall_income = gold_income * silver_per_gold + silver_income

# Build DataFrame
df = pd.DataFrame({
    "Income_Gold_Falcons": np.round(gold_income, 3),
    "Income_Silver_Falcons": np.round(silver_income, 1),
    "Overall_Income": np.round(overall_income, 2)
})

df
```

```python
plt.boxplot(df, labels=df.columns)
plt.title("Malazan Mages Income")
```
